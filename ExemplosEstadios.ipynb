```python
#Estabelecendo um ponto central do mapa
center_lat = -14.24007
center_lon = -53.18050
```

```python
#Todos os dados dos estadios na variavel consulta
consulta = list(estadios.find())
```

```python
#Converter em um Dataframe para vizualizar o mapa
df_convertido = pd.DataFrame(consulta)
```

```python
#Imprime a primeira linha do DataFrame
df_convertido.iloc[0]
```

```python
#Cria um mapa a partir do ponto central com 800x600
locationmap = folium.Map(location=[center_lat , center_lon], zoom_start=4, width=800, height=600)
```

```python
#Recuperando a quantidade de pontos
qtde = len(df_convertido)
```

```python
qtde
```

```python
#Laço de repetição para marcar cada ponto
for i in range(qtde):
  lng = float(df_convertido.iloc[i]['longitude'])
  lat = float(df_convertido.iloc[i]['latitude'])
  descripition = 'Estádio: ' + str(df_convertido.iloc[i]['nome']) + '<br> Capacidade: ' + str(df_convertido.iloc[i]['capacidade'])
  folium.Marker( [ lat , lng ],
                popup= descripition,
                 icon=folium.Icon(color='red')).add_to(locationmap)
```
